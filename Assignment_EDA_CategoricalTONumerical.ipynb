**Map Method**

```python
unique=visa_df['case_status'].unique()
d={}
p=0
for i in unique:
    d[i]=p
    p+=1
d
```

$So Now with Map method$

```python
visa_df['case_status_new']=visa_df['case_status'].map(d)
visa_df.head()
```

```python
visa_df=pd.read_csv(r"C:\Users\hp\Documents\Data Science\Naresh IT\EDA with Python\Visadataset - Visadataset.csv")
visa_df['case_status']=visa_df['case_status'].map(d)
visa_df.head()
```

```python
visa_df=pd.read_csv(r"C:\Users\hp\Documents\Data Science\Naresh IT\EDA with Python\Visadataset - Visadataset.csv")
d1={}
for i in visa_df.select_dtypes(include='object').columns:
    unique=visa_df[i].unique()
    p=0
    for j in unique:
        d1[j]=p
        p+=1
    visa_df[i]=visa_df[i].map(d1)
visa_df
```
